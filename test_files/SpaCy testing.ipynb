# WORD EMBEDDINGS

```python
import numpy as np

nlp_da_md = spacy.load("da_core_news_md")
```

```python
doc1 = nlp_da_md(text)
```

```python
# Trying to get all words that have similar meaning
your_word = "bager"

ms = nlp_da_md.vocab.vectors.most_similar(
    np.asarray([nlp_da_md.vocab.vectors[nlp_da_md.vocab.strings[your_word]]]), n=10)
words = [nlp_da_md.vocab.strings[w] for w in ms[0][0]]
distances = ms[2]
print(words)
```
